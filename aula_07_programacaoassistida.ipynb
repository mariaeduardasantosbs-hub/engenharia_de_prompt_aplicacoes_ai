```python
# Define uma lista de números inteiros
numeros = [1, 2, 3, 4, 5]
# Inicializa a variável 'soma' com zero para armazenar a soma dos números
soma = 0
# Inicializa a variável 'i' com zero, que será usada como índice para percorrer a lista
i = 0

# Inicia um loop 'while' que continua enquanto 'i' for menor que o número de elementos na lista 'numeros'
while i < len(numeros):
  # Adiciona o número atual da lista (numeros[i]) à variável 'soma'
  soma += numeros[i]
  # Incrementa o contador 'i' para passar para o próximo elemento da lista
  i += 1

# Após o loop, imprime a soma total dos números usando uma f-string para formatar a saída
print(f"A soma dos números é: {soma}")
```
